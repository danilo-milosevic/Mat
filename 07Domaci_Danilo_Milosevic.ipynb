а) Показати да је за Фробенијусову норму важи $\|A\otimes B\|_F=\|A\|_F\|B\|_F.$ Направити нумерички експеримент овог тврђења.

(5 поена)

\begin{align*}
\|A\otimes B\|_F &= tr((A\otimes B)^T(A\otimes B))\\
&= tr((A^T\otimes B^T)(A\otimes B)) \\
&= tr((A^TA)\otimes(B^TB)) \\
&= tr(A^TA)tr(B^TB) \\
&= \|A\|_F \|B\|_F \\
\end{align*}

````python
A = rndm.randn(3, 4)
B = rndm.randn(2, 5)
K = np.kron(A, B)
lhs = np.linalg.norm(K, 'fro')
rhs = np.linalg.norm(A, 'fro') * np.linalg.norm(B, 'fro')
print(f"||AxB||_F = {lhs:.6f}")
print(f"||A||_F * ||B||_F = {rhs:.6f}")
print(f"Jednaki: {np.isclose(lhs, rhs)}")
````

б) Реализовати функцију у Python-у која решава овај проблем **без експлицитног формирања Катри–Рао производа**.  
Упоредити меморијско заузеће вашег приступа у односу на нативни приступ са `np.kron`.

**Помоћ:** За ефикасно израчунавање десне стране
$$
z = (A \odot B)^{T} y,
$$
приметити да ако вектор $y$ реструктурирамо у матрицу $Y \in M_{p \times m}$ (тако да је $y = vec(Y)$),  
тада је $i$-ти елемент вектора $z$ једнак
$$
z_i = b_i^{T} \, Y \, a_i,
$$
где су $a_i$ и $b_i$ колоне матрица $A$ и $B$.

У NumPy-ју се овај вектор може добити **без петљи**, коришћењем матричног множења и сумирања дуж одређене осе, или преко:
```python
np.einsum('pi,pm,mi->i', B, Y, A)
```

(10 поена)

````python
def solve_khatri_rao_ls(A, B, y):
    m, n = A.shape
    p, _ = B.shape
    
    ATA = A.T @ A
    BTB = B.T @ B
    M = ATA * BTB
    
    Y = y.reshape(p, m)
    z = np.einsum('pi,pm,mi->i', B, Y, A)
    mem = (M.nbytes + z.nbytes)/1024
    
    x = np.linalg.solve(M, z)
    return x, mem

m, p, n = 100, 80, 20
A = rndm.randn(m, n)
B = rndm.randn(p, n)
x_true = rndm.randn(n)

C= np.column_stack([np.kron(A[:, j], B[:, j]) for j in range(n)])
y = C @ x_true

x_eff, mem_eff = solve_khatri_rao_ls(A, B, y)
x_naive= np.linalg.lstsq(C, y, rcond=None)[0]

mem_full = C.nbytes/1024 
print(f"Naivna metoda {mem_full:.2f} KB")
print(f"Efikasna metoda {mem_eff:.2f} KB")
````
