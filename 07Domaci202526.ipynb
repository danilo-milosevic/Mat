б) Реализовати функцију у Python-у која решава овај проблем **без експлицитног формирања Катри–Рао производа**.  Упоредити меморијско заузеће вашег приступа у односу на приступ са `np.kron`.

**Помоћ:** За ефикасно израчунавање десне стране
$$
z = (A \odot B)^{T} y,
$$
приметити да ако вектор $y$ реструктурирамо у матрицу $Y \in M_{p \times m}$ (тако да је $y = vec(Y)$), тада је $i$-ти елемент вектора $z$ једнак
$$
z_i = b_i^{T} \, Y \, a_i,
$$
где су $a_i$ и $b_i$ колоне матрица $A$ и $B$.

У NumPy-ју се овај вектор може добити **без петљи**, коришћењем матричног множења и сумирања дуж одређене осе, или преко:
```python
np.einsum('pi,pm,mi->i', B, Y, A)
```
